```python
import qiskit
import numpy as np
import matplotlib.pyplot as plt
from qiskit_aer.primitives import Sampler
```

```python
qreg = qiskit.QuantumRegister(3, name="z")
#creg = qiskit.ClassicalRegister(1, name="creg")
qc = qiskit.QuantumCircuit(qreg)#, creg)
qc.h(qreg[2])
qc.crz(np.pi/2, qreg[1], qreg[2])
qc.crz(np.pi/4, qreg[0], qreg[2])
qc.h(qreg[1])
qc.crz(np.pi/2, qreg[0], qreg[1])
qc.h(qreg[0])
qc.swap(qreg[2], qreg[0])

qc.draw("mpl", initial_state=True, reverse_bits=True)
```

```python
qreg = qiskit.QuantumRegister(4, name="z")
creg = qiskit.ClassicalRegister(1, name="result")
qc = qiskit.QuantumCircuit(qreg)#, creg)

qc.h(qreg[3])
qc.crz(np.pi/2, qreg[0], qreg[3])
qc.crz(np.pi/2, qreg[1], qreg[3])
qc.crz(np.pi/2, qreg[2], qreg[3])
qc.cx(qreg[0], qreg[2])
qc.cx(qreg[1], qreg[2])
qc.crz(-np.pi/2, qreg[2], qreg[3])
qc.h(qreg[3])


qc.draw("mpl", initial_state=True)
```

```python
qreg = qiskit.QuantumRegister(4, name="z")
qreg_result = qiskit.QuantumRegister(1, name="y_2")
creg = qiskit.ClassicalRegister(1, name="result")
qc = qiskit.QuantumCircuit(qreg, qreg_result)#, creg)

qc.ccx(qreg[0], qreg[1], qreg_result)
qc.ccx(qreg[1], qreg[2], qreg_result)
qc.ccx(qreg[2], qreg[3], qreg_result)
qc.ccx(qreg[0], qreg[2], qreg_result)
qc.ccx(qreg[0], qreg[3], qreg_result)
qc.ccx(qreg[1], qreg[3], qreg_result)

qc.draw("mpl")
```
